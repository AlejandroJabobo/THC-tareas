```python
x = np.linspace(0, 6 * np.pi, num=100)
y = np.sin(x)
z= np.cos(x)

plt.subplot(1,2,1)
plt.plot(x,y, '*-g',mec='k')
plt.title("seno")

plt.subplot(1,2,2)
plt.plot(x,z,'p-.b', mec='r')
plt.title("coseno")


plt.suptitle("Graficas")
plt.show
```

```python
def graficacion(x):
  z = np.cos(x)
  y = np.sin(x)
  plt.subplot(1,2,1)
  plt.plot(x,y, '*-g',mec='k')
  plt.title("seno")

  plt.subplot(1,2,2)
  plt.plot(x,z,'p-.b', mec='r')
  plt.title("coseno")

x = np.linspace(0, 6 * np.pi, num=1000)
graficacion(x)
```
