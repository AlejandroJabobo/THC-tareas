```python
x = np.linspace(0, 6 * np.pi, num=100)
y = np.sin(x)
z= np.cos(x)

plt.plot(y, '*-g',mec='k')
plt.plot(z, 'p-.b', mec='r')

plt.xlabel("eje x")
plt.ylabel("eje y")

plt.show
```

```python
def graficacion(x):
  y = np.cos(x)
  z = np.sin(x)
  plt.plot(y)
  plt.plot(z)
x = np.linspace(0, 6 * np.pi, num=1000)
graficacion(x)
```
