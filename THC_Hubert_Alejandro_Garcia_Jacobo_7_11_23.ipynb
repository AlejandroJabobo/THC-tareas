```python
Dates=SL.index[0:-1:10]
NCC= SL['No. of confirmed cases' ][0:-1:10]
NCD=SL['No. of confirmed deaths'][0:-1:10]
```

```python

plt.subplot(2,1,1)
plt.plot(Dates,NCC,'.-g')
plt.title("Infectados")
plt.ylabel('Casos confirmados')
plt.xlabel('Periodo del tiempo')
plt.xticks(rotation=90, fontsize=8)

plt.subplot(2,1,2)
plt.plot(Dates,NCD, '.-k')
plt.title('Decesos')
plt.suptitle("  Avence del ebola en Sierra Leona")
plt.ylabel('Casos confirmados')
plt.xlabel('Periodo del tiempo')
plt.xticks(rotation=90, fontsize=8)
plt.show
```

```python
PIvsD=100*SL['No. of confirmed cases' ]/SL['No. of confirmed deaths']
SL['PIvsD']=PIvsD
SL.head()
```

```python
recuperados=(100-PIvsD/100)*SL['No. of confirmed cases' ]
SL['recuperados']=recuperados
SL.head()
```
